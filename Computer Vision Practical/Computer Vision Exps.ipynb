```python
# Program 1
sub_1 = int(input("Enter the marks of subject 1: "))
sub_2 = int(input("Enter the marks of subject 2: "))
sub_3 = int(input("Enter the marks of subject 3: "))
sub_4 = int(input("Enter the marks of subject 4: "))
sub_5 = int(input("Enter the marks of subject 5: "))

percentage = (sub_1+sub_2+sub_3+sub_4+sub_5)*100/500

print(f'Student Percentage: {percentage:.2f}%')

if percentage < 33:
    print("Failed!")
elif percentage >= 33 and percentage <= 50:
    print("E Grade!")
elif percentage > 50 and percentage <= 60:
    print("D Grade!")
elif percentage > 60 and percentage <= 70:
    print("C Grade!")
elif percentage > 70 and percentage <= 80:
    print("B Grade!")
elif percentage > 80 and percentage <= 90:
    print("A Grade!")
else:
    print("A+ Grade")
```

```python
# Program 2
num = int(input('Enter the number: '))
if num > 1:
    for i in range(2,int(num/2)+1):
        if (num%i) == 0:
            print(num,'is not a prime number')
            break
    else:
        print(num,'is a prime number')
else:
    print(num,'is not a prime number')
```

```python
# Program 3
def recursive_sum(lst):
    if not lst:
        return 0
    else:
        return lst[0] + recursive_sum(lst[1:])
my_list = []
for i in range(1,6):
    num = int(input(f'Enter the element {i} in list:'))
    my_list.append(num)
result = recursive_sum(my_list)
print(f'Sum of elements present in list:{result}')
```

```python
# Program 4
num = int(input('Enter the number:'))
factorial = 1
if num < 0:
    print('Sorry, factorial does not exist for negative numbers')
elif num == 0:
    print('The factorial of 0 is 1')
else:
    for i in range(1,num+1):
        factorial = factorial*i
    print(f'The factorial of {num} is {factorial}')
```

```python
# Program 5
list_of_nums = []
for i in range(1,6):
    nums = float(input(f'Enter the {i} number:'))
    list_of_nums.append(nums)
avg = sum(list_of_nums)/len(list_of_nums)
print(f'Average of the five numbers taken from the user is {avg}')
```

```python
# Program 13

import cv2 
video_path = 'sample-5s.mp4' 
cap = cv2.VideoCapture(video_path) 
if not cap.isOpened(): 
    print("Error: Could not open video.") 
    exit() 
while cap.isOpened(): 
 # Capture frame-by-frame 
 ret, frame = cap.read() 
 # If frame is read correctly ret is True 
    if ret: 
        cv2.imshow(‘Video’,frame) 
        if cv2.waitKey(25) & 0xFF == ord('q'): 
            break 
    else: 
        break 
cap.release() 
# Close all the frames 
cv2.destroyAllWindows()
```

```python
# Program 14

import cv2 
import os 
video_path = 'sample.mp4' 
cap = cv2.VideoCapture(video_path) 
if not cap.isOpened(): 
    print("Error: Could not open video.") 
    exit() 
output_dir = 'Extracted_Images' 
os.makedirs(output_dir, exist_ok=True) 
frame_count = 0 
while True: 
    ret, frame = cap.read() 
    if not ret: 
        break 
    frame_filename = os.path.join(output_dir, f"Image_{frame_count:04d}.jpg") 
    cv2.imwrite(frame_filename, frame)
    frame_count += 1 
# Release the VideoCapture object 
cap.release() 
print(f"{frame_count} images extracted and saved in '{output_dir}' directory.") 
```

```python
# Program 15

# 1. Scaling (Resizing) Image
import cv2 
import matplotlib.pyplot as plt 
img = cv2.imread('ship.jpg') 
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) 
resized_img = cv2.resize(img, (400, 300)) 
plt.imshow(resized_img)

# 2. Translation (Shifting) an Image
import cv2 
import numpy as np 
img = cv2.imread('ship.jpg') 
rows, cols, channels = img.shape 
M = np.float32([[1, 0, 100], [0, 1, 50]]) 
translation_img = cv2.warpAffine(img, M, (cols, rows)) 
plt.imshow(translation_img)

# 3. Rotation of image
import cv2 
img = cv2.imread('ship.jpg', cv2.IMREAD_GRAYSCALE) 
rows, cols = img.shape 
M = cv2.getRotationMatrix2D((cols/2, rows/2), 90, 1) 
rotated_img = cv2.warpAffine(img, M, (cols, rows)) 
plt.imshow(rotated_img)

# 4. Perspective Transformation
import cv2 
import numpy as np 
img = cv2.imread('ship.jpg') 
pts1 = np.float32([[100, 100], [566, 100], [100, 900], [566, 900]]) 
width, height = 400, 600 
pts2 = np.float32([[0, 0], [width, 0], [0, height], [width, height]]) 
M = cv2.getPerspectiveTransform(pts1, pts2) 
transformed_img = cv2.warpPerspective(img, M, (width, height)) 
plt.imshow(transformed_img)
```

```python
# Program 16
import cv2 
import numpy as np 
img1 = cv2.imread(r”C:\Users\Admin\Desktop\CV\lion.jpg”) 
img2 = cv2.imread(r”C:\Users\Admin\Desktop\CV\cub.jpg”) 
sift = cv2.xfeatures2d.SIFT_create() 
kp1, des1 = sift.detectAndCompute(img1, None) 
kp2, des2 = sift.detectAndCompute(img2, None) 
matcher = cv2.BFMatcher() 
matches = matcher.match(des1, des2) 
pts1 = np.float32([kp1[m.queryIdx].pt for m in matches]).reshape(-1, 1, 2) 
pts2 = np.float32([kp2[m.trainIdx].pt for m in matches]).reshape(-1, 1, 2) 
H, _ = cv2.findHomography(pts1, pts2, cv2.RANSAC, 5.0) 
aligned_img = cv2.warpPerspective(img1, H, (img2.shape[1], img2.shape[0])) 
cv2.imshow('Aligned Image', aligned_img) 
29 
cv2.waitKey(0) 
cv2.destroyAllWindows()
```
